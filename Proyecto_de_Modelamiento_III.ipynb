```
def color_rep(r,g,b):  
  cr=[]; cg=[]; cb=[]
  fig,axes = plt.subplots(1,3, figsize=(14,4)); p=25
  r = axes[0].hist(r, bins=p,color='maroon')
  lr0 = list(r[0]); lr1 = list(r[1]); lro = sorted(list(r[0]),reverse=True)
  g = axes[1].hist(g, bins=p, color='seagreen')
  lg0 = list(g[0]); lg1 = list(g[1]); lgo = sorted(list(g[0]),reverse=True)
  b = axes[2].hist(b, bins=p, color='royalblue')
  lb0 = list(b[0]); lb1 = list(b[1]); lbo = sorted(list(b[0]),reverse=True); plt.clf()
  for i in lro:
    ind = lr0.index(i); cr.append(round((lr1[ind]+lr1[ind+1])/2))
  for i in lgo:
    ind = lg0.index(i); cg.append(round((lg1[ind]+lg1[ind+1])/2))
  for i in lbo:
    ind = lb0.index(i); cb.append(round((lb1[ind]+lb1[ind+1])/2))
  RGB = np.array([[[cr[0],cg[0],cb[0]]]])
  return RGB
```
